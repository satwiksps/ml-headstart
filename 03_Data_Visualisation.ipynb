## Load Dataset
Weâ€™ll use Pandas to load the Pima Indians Diabetes dataset directly from a GitHub URL.

```python
from pandas import read_csv

url = 'https://raw.githubusercontent.com/erojaso/MLMasteryEndToEnd/master/data/pima-indians-diabetes.data.csv'
column_names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']

# Load the CSV into a DataFrame
data = read_csv(url, names=column_names)
```

## Histogram Plots
Histograms show the distribution of individual features. This helps in understanding skewness and outliers.

```python
from matplotlib import pyplot

# Plot a histogram for each feature
data.hist()
pyplot.suptitle("Histograms of Features", fontsize=14)
pyplot.show()
```

## Density Plots
Density plots are smooth curves representing the distribution. These are helpful for identifying skewness and distribution shape.

```python
# Plot density curves for each feature
data.plot(kind='density', subplots=True, layout=(3,3), sharex=False, figsize=(10,8))
pyplot.suptitle("Density Plots of Features", fontsize=14)
pyplot.show()
```

## Box and Whisker Plots
Box plots summarize data using median, quartiles, and outliers.

```python
# Box plots for visualizing median and outliers
data.plot(kind='box', subplots=True, layout=(3,3), sharex=False, sharey=False, figsize=(10,8))
pyplot.suptitle("Box and Whisker Plots", fontsize=14)
pyplot.show()
```

## Correlation Matrix (Text + Heatmap)
Correlation shows how strongly features are related to each other. High correlations may indicate redundancy.

```python
from pandas import set_option

# Configure Pandas display settings
set_option('display.width', 100)
set_option('display.precision', 3)

# Calculate Pearson correlation coefficients
correlations = data.corr(method='pearson')
print(correlations)
print()
```

## Correlation Heatmap (Visual)
A heatmap gives a quick visual representation of correlation strengths.


```python
from matplotlib import pyplot as plt

# Basic correlation heatmap
plt.matshow(correlations, cmap='coolwarm')
plt.title("Correlation Matrix", pad=20)
plt.colorbar()
plt.show()
```

## Scatter Plot Matrix
Scatter matrices are useful to observe pairwise relationships between all numerical variables.

```python
from pandas.plotting import scatter_matrix

# Pairwise scatter plots
scatter_matrix(data, figsize=(10, 10), diagonal='hist')
pyplot.suptitle("Scatter Matrix", fontsize=14)
pyplot.show()
```
